```python
import numpy as np
u1i1 = 1
u1i2 = 1
u2i1 = 5
u2i2 = 5
(u1i1*u2i1+u1i2*u2i2)/((u1i1**2+u1i2**2)**0.5*(u2i1**2+u2i2**2))**0.5
```

```python
import numpy as np
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
# 固定 User2 item 1 及 item2 的評分 為 1
u2i1 = 10
u2i2 = 10

# 測試 User1: item1 及 item2 的評分
u1i1 = np.arange(1, 6, 1)
u1i2 = np.arange(1, 6, 1)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
X, Y = np.meshgrid(u1i1, u1i2)
R = (X*u2i1+Y*u2i2)/((X**2+Y**2)**0.5*(u2i1**2+u2i2**2))**0.5

# Plot the surface.
surf = ax.plot_surface(X, Y, R, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# 產生bar
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
print(R)
```
